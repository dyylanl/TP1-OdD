```python
import pandas as pd
```

```python
logs = pd.read_csv('logs.csv',usecols=['contributor_id','action'])
```

```python
lang = pd.read_csv('languages.csv')
```

```python
merg = logs.merge(lang,left_on='contributor_id',right_on='babel_user',how='outer')
```

```python
df = merg.loc[merg.action == 'thank']
```

```python
df['dif'] = 1
df
```

```python
df=df.groupby(['contributor_id','babel_lang']).agg({'dif':'count'})
df
```
